Theory of the Lab Session --

Confusion Matrix:

  TP = True Positive

  TN = True Negative

  FP = False Positive

  FN = False Negative


Accuracy = (TP + TN) / (TP + FN + FP + TN)

Precision = TP / (TP + FP)

Recall = TN / (TN + FP)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/Social_Network_Ads.csv')

data
```

```python
data.info
```

```python
x = data.drop(['Purchased', 'User ID', 'Gender'], axis = 1)
y = data['Purchased']

x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_trian = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
final = pd.DataFrame(np.c_[y_test, y_pred], columns = ['Actual', 'Predicted'])
final
```

```python
# The correct import statement should be:
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
```
